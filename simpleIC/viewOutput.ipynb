```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import pynbody as pb
import glob as gl
import natsort as ns
```

```python
files = ns.natsorted(gl.glob('*.[0-9]*[0-9]'))
print(files)
```

```python
xmin, xmax = -2.0, 2.0
ymin, ymax = -2.0, 2.0

for idx, val in enumerate(files):
    snap = pb.load(val)
    time = idx+1
    
    fig, axes = plt.subplots(figsize=(8,8))
    for idx1 in range(len(snap)):
        circle = plt.Circle((snap['pos'][idx1][0], snap['pos'][idx1][1]), snap['eps'][idx1]*2, color='r')
        axes.add_artist(circle)
        arrow = plt.arrow(snap['pos'][idx1][0], snap['pos'][idx1][1], snap['vel'][idx1][0], snap['vel'][idx1][1], head_width=0.1, head_length=0.1)
        axes.add_artist(arrow)
        axes.set_xlim(xmin, xmax)
        axes.set_ylim(ymin, ymax)
        axes.set_xlabel('X position')
        axes.set_xlabel('Y position')
        axes.set_title('T = ' + str(time))
```
